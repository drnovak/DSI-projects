```python
df.columns
```

```python
#rename some columns
df.rename(columns={'artist.inverted': 'artist','date.entered': 'date_enter', 
                   'date.peaked': 'date_peak'}, inplace=True)
df.head(2)
```

```python
#convert dates(type) from string to datetime
df['date_enter'] = pd.to_datetime(df['date_enter'])
df['date_peak'] = pd.to_datetime(df['date_peak'])
```

```python
df['date_enter'].head()
df.head(2)
```

```python
# df['year'] can be dropped because of new datetime columns
df = df.drop('year', axis=1)
df.head(2)
```

```python
df['time'].describe()
```

```python
#create function to convert time into #of seconds
def timeSplit(string):
    temp = string.split(":")
    minutes = int(temp[0])
    seconds = int(temp[1])
    return minutes*60 + seconds

df['time'] = df['time'].apply(timeSplit)
```

```python
df.head(2)
```

```python
df.time.describe()
```

```python
#rename 'xNweek' to 'weekN' format
week_columns = []
counter = 1
for column in df.columns:
    if column[0] == 'x':
        week_columns.append('week'+ str(counter))
        counter += 1
    else:
        week_columns.append(column)

week_columns
df.columns = week_columns
```

```python
df.head(2)
```

```python
df.dtypes.head(10)
```

```python
df.week1 = [float(i) for i in df.week1]
```

##### Using Pandas' built in `melt` function, pivot the weekly ranking data to be long rather than wide. As a result, you will have removed the 72 'week' columns and replace it with two: Week and Ranking. There will now be multiple entries for each song, one for each week on the Billboard rankings.

```python
df.head(2)
```

```python
# Step 1 - Separate into columns containing song information and columns with week ranking information.

week_columns = []
info_columns = []
for col in df:
    if col[0] == 'w':
        week_columns.append(col)
    else:
        info_columns.append(col)

print week_columns
print '\n'
print info_columns
```

```python
# pd.melt creates a new (melted) dataframe. Each song will have a row for per weekly ranking.

melted_df = pd.melt(df, id_vars=info_columns, value_vars=week_columns, var_name='week', value_name='ranking')
melted_df.head(10)
```

```python
melted_df.shape
```

```python
melted_df.describe()
```

```python
# Drop all rows that contain NaNs in ranking column

melted_df_no_nan = melted_df.dropna()
melted_df_no_nan
```

```python
melted_df_no_nan.shape
```

```python
melted_df_no_nan['artist'].value_counts().head(15)
```

```python
melted_df_no_nan['ranking'].mean()
```

```python
melted_df_no_nan['time'].mean()
```

```python
bb_data = pd.pivot_table(melted_df_no_nan,
                             index=['artist', 'track', 'time', 'genre', 'date_enter', 'date_peak', 'week'],
                             values=['ranking'])
bb_data.head(10)
```
